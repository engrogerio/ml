```python
import matplotlib.pyplot as plt
from scipy.stats import stats
import numpy as np
from sklearn.metrics import r2_score
```

```python
# internet velocity x hour of the day
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [101,89,91,60,60,53,58,65,70,69,75,76,77,79,90,99,99,100]
print(len(x), len(y))
```

```python
slope, intercept, r, p, std_error = stats.linregress(x, y)
```

```python
def find_new_y(x):
    return slope * x + intercept
```

```python
# Plot scatter with regression line
y_values = list(map(find_new_y, x))

plt.scatter(x, y, label='original data')
plt.plot(x, y_values, 'r', label='fitted line')
plt.legend()
print(r)
```

```python
poly_model = np.poly1d(np.polyfit(x, y, 2))
# generate evenly spaced 100 numbers for interval 1-22
poly_line = np.linspace(1, 22, 100)
plt.scatter(x, y, label='original data')
plt.plot(poly_line, poly_model(poly_line), 'r', label='fitted_curve')
plt.legend()
print('r-squared = ', (r2_score(y, poly_model(x)))) 
print('1 - 0.7 = strong correlation')
print('0.4 - 0.7 = moderate')
print('0.2 - 0.4 = weak')
print('0 - 0.2 = no correlation')
```
